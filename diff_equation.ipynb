```python
import numpy as np
import matplotlib.pyplot as plt

# Definisi fungsi dy/dx
def f(x, y):
    return -2 * x * y

# Nilai awal dan langkah
x0, y0 = 0, 1
h = 0.1
x_end = 2

# Membuat array untuk x dan y
x_values = np.arange(x0, x_end + h, h)
y_values = np.zeros(len(x_values))
y_values[0] = y0

# Metode Euler
for i in range(1, len(x_values)):
    x_prev = x_values[i-1]
    y_prev = y_values[i-1]
    y_values[i] = y_prev + h * f(x_prev, y_prev)

# Plot hasil metode Euler
plt.plot(x_values, y_values, label='Euler (Numerik)', marker='o')

# Solusi analitik
def analytical_solution(x):
    return np.exp(-x**2)

y_exact = analytical_solution(x_values)
plt.plot(x_values, y_exact, label='Solusi Analitik', linestyle='--')

# Label dan tampilan plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Metode Euler untuk dy/dx = -2xy')
plt.legend()
plt.grid()
plt.show()
```
